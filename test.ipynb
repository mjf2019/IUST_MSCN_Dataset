```python
import pandas as pd
import os

def process_csv_files(file1, file2):
    """
    دو فایل CSV را خوانده، دو ستون آخر را حذف کرده و با همان نام ذخیره می‌کند
    """
    # خواندن فایل‌های CSV
    df1 = pd.read_csv(file1)
    df2 = pd.read_csv(file2)
    
    # بررسی اینکه حداقل دو ستون وجود داشته باشد
    if len(df1.columns) >= 2:
        df1 = df1.iloc[:, :-2]  # حذف دو ستون آخر
    else:
        print(f"توجه: فایل {file1} کمتر از دو ستون دارد")
    
    if len(df2.columns) >= 1:
        df2 = df2.iloc[:, :-1]  # حذف دو ستون آخر
    else:
        print(f"توجه: فایل {file2} کمتر از دو ستون دارد")
    
    # ذخیره فایل‌ها با همان نام
    df1.to_csv(file1, index=False)
    df2.to_csv(file2, index=False)
    
    print(f"فایل‌ها با موفقیت پردازش و ذخیره شدند:")
    print(f"- {file1}: {len(df1.columns)} ستون باقی‌مانده")
    print(f"- {file2}: {len(df2.columns)} ستون باقی‌مانده")

# مثال استفاده:
# نام فایل‌های خود را اینجا وارد کنید
file1 = 'DATASETS/CDR-MLC/scale_0.001/Short/CDR_MLC_Shuffle.csv'  # نام فایل اول
file2 = "DATASETS/CDR-MLC/scale_0.001/Long/CDR_MLC_Shuffle.csv"  # نام فایل دوم

# اجرای تابع
process_csv_files(file1, file2)
```
