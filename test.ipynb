```python
import matplotlib.pyplot as plt
import numpy as np

# Data for ISCX-VPN and ISCX-TOR
models = ['DISTILLER', 'DeepPacket', 'GRAIN', 'FlowPrint', 'Proposed RF', 'CDR-MLC']
accuracy_vpn = [0.9319, 0.9641, 0.8511, 0.8721, 0.8792, 0.9973]
f1_vpn = [0.8938, 0.9589, 0.8651, 0.7632, 0.8564, 0.9916]
precision_vpn = [0.8775, 0.9566, 0.8781, 0.7643, 0.8704, 0.9875]
recall_vpn = [0.9107, 0.9611, 0.8526, 0.8243, 0.9257, 0.9926]

accuracy_tor = [0.9338, 0.9276, 0.7361, 0.6307, 0.9681, 0.9727]
f1_tor = [0.8605, 0.9250, 0.7654, 0.7784, 0.9423, 0.9700]
precision_tor = [0.8297, 0.9251, 0.8020, 0.9293, 0.9576, 0.9757]
recall_tor = [0.8937, 0.9249, 0.7321, 0.6697, 0.9522, 0.9743]

x = np.arange(len(models))  # the label locations
width = 0.2  # the width of the bars

# Create subplots for both ISCX-VPN and ISCX-TOR
fig, ax = plt.subplots(2, 1, figsize=(12, 10), sharex=True)

# Plot for ISCX-VPN
ax[0].bar(x - 1.5 * width, accuracy_vpn, width, label='Accuracy', color='blue')
ax[0].bar(x - 0.5 * width, f1_vpn, width, label='F1-score', color='green')
ax[0].bar(x + 0.5 * width, precision_vpn, width, label='Precision', color='orange')
ax[0].bar(x + 1.5 * width, recall_vpn, width, label='Recall', color='red')
ax[0].set_title('ISCX-VPN Dataset')
ax[0].set_ylabel('Scores')
ax[0].legend()

# Plot for ISCX-TOR
ax[1].bar(x - 1.5 * width, accuracy_tor, width, label='Accuracy', color='blue')
ax[1].bar(x - 0.5 * width, f1_tor, width, label='F1-score', color='green')
ax[1].bar(x + 0.5 * width, precision_tor, width, label='Precision', color='orange')
ax[1].bar(x + 1.5 * width, recall_tor, width, label='Recall', color='red')
ax[1].set_title('ISCX-TOR Dataset')
ax[1].set_ylabel('Scores')

# Labels, title and adjustments
ax[1].set_xticks(x)
ax[1].set_xticklabels(models)
plt.xticks(rotation=45)
plt.tight_layout()

# Display chart
plt.show()
```
